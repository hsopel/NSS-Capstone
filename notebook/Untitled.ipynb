```python
import pandas as pd
```

```python
con_vs_cap = pd.read_csv('../data/wind-energy-consumption-vs-installed-wind-energy-capacity.csv')
```

```python
con_vs_cap = con_vs_cap.iloc[2:]
```

```python
con_vs_cap = con_vs_cap.reset_index(drop=True)
```

```python
con_vs_cap
```

```python
share = pd.read_csv('../data/wind-share-energy.csv')
```

```python
share
```

```python
world_combined = pd.merge(con_vs_cap, share, on = ['Code', 'Entity', 'Year'])
```

```python
world_combined = world_combined[world_combined['Year'] > 1979]
```

```python
world_combined = world_combined.drop(columns = 'Continent')
```

```python
continents = ['North America', 'South America', 'Africa', 'Europe', 'Asia', 'World']
```

```python
world_combined = world_combined[~world_combined['Entity'].isin(continents)]
```
